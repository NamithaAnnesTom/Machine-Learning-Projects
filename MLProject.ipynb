***Extracting and listing top 100 most frequent values from five different columns-'location','rest_type','dish_liked','cuisines','listed_in(type)' of the dataset***

```python
top_locations=df['location'].value_counts()[:100].index.tolist()
top_rest_types=df['rest_type'].value_counts()[:100].index.tolist()
top_dishes=df['dish_liked'].value_counts()[:100].index.tolist()
top_cuisines=df['cuisines'].value_counts()[:100].index.tolist()
top_types=df['listed_in(type)'].value_counts()[:100].index.tolist()
```

***This code defines a function clean_categorical_columns that takes two parameters value,common_values. Categorizing values in these categorical columns by setting missing values as 'None', keeps value that matches common_values list unchanged and categorize all other values as other***

```python
def clean_categorical_columns(value,common_values):
  if pd.isnull(value):
    return None
  elif value in common_values:
    return value
  else:
    return 'other'
```

***Using clean_categorical_columns function we replace values that are not common with the label 'other' and values that are common are kept as they are.The common values are defined by the lists, top_locations,top_rest_types,top_cuisines,top_types.This approach is useful for consolidating less frequent values into a common category and improving the interpretability of categorical data.***

```python
df['location']=df['location'].apply(clean_categorical_columns,common_values=top_locations)
df['rest_type']=df['rest_type'].apply(clean_categorical_columns,common_values=top_rest_types)
df['dish_liked']=df['dish_liked'].apply(clean_categorical_columns,common_values=top_dishes)
df['cuisines']=df['cuisines'].apply(clean_categorical_columns,common_values=top_cuisines)
df['listed_in(type)']=df['listed_in(type)'].apply(clean_categorical_columns,common_values=top_types)
```

***MODEL CREATION***

***CLASSIFICATION MODEL: it predicts if a given restaurant is affordable or expensive based on all given input features***

***Seperating input and output data***

```python
x1=df.drop(columns=['expensive'])
y1=df['expensive']
```

***Splitting the data into training data and testing data using train_test_split function.***

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y1,random_state=42,test_size=0.30)
```

***Scaling the dataset using StandardScaler: it standardizes the features of a dataset by removing the mean and scaling it to unit variance.***

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

***Model creation using different classification algorithms:***

```python

from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
nb_model=BernoulliNB()
svm_model=SVC()
tree_model=DecisionTreeClassifier(criterion='entropy')
forest_model=RandomForestClassifier(n_estimators=100)
lst=[nb_model,svm_model,tree_model,forest_model]
```

***MODEL EVALUATION:***

***DISPLAYING CLASSIFICATION REPORT AND CONFUSION MATRIX:***

***CLASSIFICATION REPORT-the classification report is a useful tool for evaluating the performance of a classification model. It provides a comprehensive summary of various evaluation metrics for each class.The classification reports typically includes the following metrics:***

***Precision: the proportion of true positive predictions out of all positive predictions.***

***Recall: the proportion of true positive predictions out of all positive instances.***

***F1-score: the harmonic mean of precision and recall.***

***Support: the number of actual occurences of the class in the test set.***

***Accuracy: the accuracy of the classifier on the entire dataset.***

***Macro avg: the average of precision, recall and F1-score across all classes, giving each class equal weight.***

***Weighted avg: the average of precision,recall and F1-score across all classes, weighted by the number of instances in each class.***

***CONFUSION MATRIX: it is table that is often used to describe the performance of a classification model on a set of test data for which the true values are known, it helps to understand the distribution of predictions for each class and provides insights into how well the model is performing for different classes.***

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,ConfusionMatrixDisplay
for i in lst:
  print(i)
  print("*****************************")
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print(confusion_matrix(y_test,y_pred))
  print(classification_report(y_test,y_pred))
  labels=['Expensive','Affordable']
  result=confusion_matrix(y_test,y_pred)
  cmd=ConfusionMatrixDisplay(result,display_labels=labels)
  cmd.plot()
```

***The best output is obtained for models which are deployed using Rando Forest and Decision Tree algorithms***

***REGRESSION MODEL: to predict the approx_cost(for two people) at a restaurant based on given input features***

***Repeat all preprocessing steps of classification model***

```python
x2=df.drop(columns=['approx_cost(for two people)'])
y2=df['approx_cost(for two people)']
```

```python
from sklearn.model_selection import train_test_split
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x2_train)
x2_train=scaler.fit_transform(x2_train)
x2_test=scaler.transform(x2_test)
```

***Using various regression algorithms for prediction.***

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
linear_model = LinearRegression()
tree_regmodel = DecisionTreeRegressor()
forest_regmodel=RandomForestRegressor()
reg_lst=[linear_model,tree_regmodel,forest_regmodel]
```

***Displaying evaluation metrics of regression model:***

***r2_score- it represents the proportion of the variance for the dependent variable that's explained by independent variables in a regression model, its value ranges from 0 to 1.***

***MAPE- mean absolute percentage error, measures the percentage difference between predicted and actual values, averaged across all data points.***

```python
from sklearn.metrics import r2_score,mean_absolute_percentage_error
for i in reg_lst:
  i.fit(x2_train,y2_train)
  y2_pred=i.predict(x2_test)

  print(i,' R-squared is',(r2_score(y2_test,y2_pred)))

  print(i,'Mean Absolute Percentage Error is',(mean_absolute_percentage_error(y2_test,y2_pred)))

  plt.figure(figsize=(10, 6))
  sns.regplot(x=y2_test, y=y2_pred)
  plt.xlabel('Actual Cost')
  plt.ylabel('Predicted Cost')
  plt.title('Regression Plot: Actual vs. Predicted Cost')
  plt.show()
```

***The best r2_score is obtained for Random forest regressor followed by Decision tree regressor***
